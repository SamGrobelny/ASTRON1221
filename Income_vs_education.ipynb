```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Extract the first column and remove the first one entry.
Education_levels = np.array([
"Less than high school completion",
"High school completion",
"Some college, no degree",
"Associate’s degree",
"Bachelor’s degree",
"Master’s or higher degree"
])
```

```python
g_educ_arr = np.arange(6)
male_income_arr = np.array([36700, 46400, 50200, 54200, 75100, 89200])
female_income_arr = np.array([30900, 36700, 39100, 43100, 60800, 75700])
z_linear_male = np.polyfit(g_educ_arr,male_income_arr,1)
z_linear_female = np.polyfit(g_educ_arr,female_income_arr,1)
z_quadratic_male = np.polyfit(g_educ_arr,male_income_arr,2)
z_quadratic_female = np.polyfit(g_educ_arr,female_income_arr,2)
O_male = male_income_arr
O_female = female_income_arr
O_err_male = np.array([3110, 1070, 2000, 4130, 4110, 8910])
O_err_female = np.array([3310, 960, 3070, 2580, 860, 4940])
C_linear_male = z_linear_male[0] * g_educ_arr + z_linear_male[1]
C_quadratic_male = z_quadratic_male[0] * g_educ_arr**2 + z_quadratic_male[1] * g_educ_arr + z_quadratic_male[2]
C_linear_female = z_linear_female[0] * g_educ_arr + z_linear_female[1]
C_quadratic_female = z_quadratic_female[0] * g_educ_arr**2 + z_quadratic_female[1] * g_educ_arr + z_quadratic_female[2]


plt.errorbar(g_educ_arr, O_male, yerr=O_err_male, marker="o", linestyle = "None", color = "blue", label = "Male")
plt.errorbar(g_educ_arr, O_female, yerr=O_err_female, marker="o", linestyle = "None", color = "red", label = "Female")
plt.plot(g_educ_arr,C_linear_male, color = "blue", label = "Linear fit for male")
plt.plot(g_educ_arr,C_linear_female, color = "red", label = "Linear fit for female")
plt.plot(g_educ_arr,C_quadratic_male, color = "blue", label = "Quadratic fit for male")
plt.plot(g_educ_arr,C_quadratic_female, color = "red", label = "Quadratic fit for female")
plt.legend()
plt.xlabel("Education Level")
plt.ylabel("Income [$]")
plt.show()
```

```python
print(f"Male slope: {z_linear_male[0]:.2f}, Male intercept: {z_linear_male[1]:.2f}")
print(f"Female slope: {z_linear_female[0]:.2f}, Female intercept: {z_linear_female[1]:.2f}")
```
